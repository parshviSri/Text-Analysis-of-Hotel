```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

```

```python
base_url = "https://www.tripadvisor.in/Hotel_Review-g297628-d302480-Reviews-or"
urls= ['https://www.tripadvisor.in/Hotel_Review-g297628-d302480-Reviews-The_Leela_Palace_Bengaluru-Bengaluru_Bangalore_District_Karnataka.html']

for page_number in range(1, 200): 
    current_url = f"{base_url}{page_number * 10}-The_Leela_Palace_Bengaluru-Bengaluru_Bangalore_District_Karnataka.html#REVIEWS"
    urls.append(current_url)
print(urls)
df = pd.DataFrame(columns=['Title','Rating', 'Review'])
```

```python
import random
import config
```

```python
def rand_proxy():
    proxy =random.choice(["110.34.166.187","187.86.129.132","216.107.129.72","116.58.224.33","197.234.13.44","111.67.197.147","185.173.106.113"])
    return proxy
```

```python

def getReviews(url):
    print(url)
    safari_options = webdriver.SafariOptions()
    proxy = rand_proxy()
    safari_options.add_argument(f'--proxy-server={proxy}')
    driver = webdriver.Safari(options = safari_options)
    driver.get(url)
    row_df = pd.DataFrame(columns=['Title'])
    container = driver.find_elements(By.XPATH,"//div[@data-reviewid]")

    for element in container:
        try:
            title = element.find_element(By.XPATH,".//div[contains(@data-test-target, 'review-title')]").text
            review = element.find_element(By.XPATH,".//span[contains(@data-test-target,'review-text')]").text.replace("\n", "  ")
            #date = element.find_element(By.XPATH,".//span[contains(@class, 'ratingDate')]").get_attribute("title")
            rating = element.find_element(By.XPATH,".//span[contains(@class, 'ui_bubble_rating bubble_')]").get_attribute("class").split("_")[3]
            
            new_row_df = pd.DataFrame({"Title": [title],"Review":[review],"Rating":[rating]})
            row_df = pd.concat([row_df, new_row_df], ignore_index=True)
        except StaleElementReferenceException :
                continue
        except NoSuchElementException as e:
            print(f"Error: {e}")
            continue
    driver.quit()
    return row_df  
   
```

```python
for url in urls:
    row =getReviews(url)
    df = pd.concat([df, row], ignore_index=True)
```
